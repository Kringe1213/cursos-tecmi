```python
def preguntar():
  print("¿Comó estas?")
  print("Patron")

if __name__ == "__main__":
  print(preguntar())


def suma(num1,num2):
  return num1 + num2

def resta(num1,num2):
  return num1 - num2


def multiplicacion(num1,num2):
  return num1 * num2

def division(num1,num2):
  return num1 / num2

print(suma(2,3))
print(resta(2,3))
print(multiplicacion(2,3))
print(division(2,3))


```

```python
if __name__ == "__main__":
  print("Factorial")
  num = int(input(" Numero: "))

def Factorial(num):
  resultado = 1
  for i in range(1,num+1):
    resultado = resultado * i
  return resultado

print(Factorial(num))
```
