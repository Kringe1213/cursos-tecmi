```python
def preguntar():
  print("¿Comó estas?")
  print("Patron")

if __name__ == "__main__":
  print(preguntar())


def suma(num1,num2):
  return num1 + num2

def resta(num1,num2):
  return num1 - num2


def multiplicacion(num1,num2):
  return num1 * num2

def division(num1,num2):
  return num1 / num2

print(suma(2,3))
print(resta(2,3))
print(multiplicacion(2,3))
print(division(2,3))


```

```python
if __name__ == "__main__":
  print("Factorial")
  num = int(input(" Numero: "))

def Factorial(num):
  resultado = 1
  for i in range(1,num+1):
    resultado = resultado * i
  return resultado

print(Factorial(num))
```

```python
lista_super = []

while True:
    print("Menu de Lista de la Compra")
    print("1. Añadir artículo")
    print("2. Quitar artículo")
    print("3. Ver lista")
    print("4. Salir")

    opcion = input("Selecciona una opción (1-4): ")

    if opcion == '1':
        articulo = input("Introduce el artículo a añadir: ")
        lista_super.append(articulo)
        print(f"'{articulo}' añadido a la lista.")
    elif opcion == '2':
        if not lista_super:
            print("La lista vacía, no hay artículos para quitar.")
            continue
        articulo = input("Introduce el artículo a quitar: ")
        if articulo in lista_super:
            lista_super.remove(articulo)
            print(f"'{articulo}' eliminado de la lista.")
        else:
            print(f"'{articulo}' no se encontró en la lista.")
    elif opcion == '3':
        if lista_super:
            print("\nTu lista de la compra:")
            for i, item in enumerate(lista_super):
                print(f"{i+1}. {item}")
        else:
            print("La lista de la compra está vacía.")
    elif opcion == '4':
        print("Saliendo del programa.")
        break
    else:
        print("Por favor, elige un número del 1 al 4.")
```
