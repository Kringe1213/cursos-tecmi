```python
comidas = ["Pollo", "Pavo", "Caldo"]
comidas.append("Pizza")
comidas.insert(4, "Sushi")
comidas.remove("Caldo")
print(comidas)
comidas.pop()
comidas.sort()

print(comidas)
print(len(comidas))
for i in comidas:
  print(i)
```

```python
ingredientes = []
ingredientes.append("Leche")
ingredientes.append("Pan")
ingredientes.append("Huevos")
ingredientes.insert(0, "Manzanas")
print(ingredientes)
ingredientes.pop()
ingredientes.sort()
print(ingredientes)
print(len(ingredientes))

```

```python
print("Sistema de inscripción")
num_usuarios = int(input("Numero de usuarios: "))

usuarios = []
for i in range (num_usuarios):
  Nombre = input("Nombre de usuario: ")
  usuarios.append(Nombre)
  Nacimiento = int(input("Año de nacimiento: "))
  usuarios.append(Nacimiento)

usuarios.pop()
usuarios.pop()
usuarios.append(Nombre)
usuarios.append(Nacimiento)
print(usuarios)
```
